```python
import pandas as pd
```

```python
movies=['Kantara','bahubali','titanic','A','manjumaliBoys',]
rating=[8,9,None,7.4,8.7]
language_year=['kannada-2023','telugu-2018','english-2002','kannada-2003','malyalam-2024']
cast=[['Rishab,actress'],['prabas,anushka'],['ketevinslate,actor'],['upendra,actress'],['Subhash,actor']]
prime=['yes','yes','no','no','yes']
hotstar=['no','yes','yes','yes','no']


d={'Movies_names':movies,
   'Movie_rating':rating,
   'Language_year':language_year,
   'Movie_cast':cast,
   'prime_ott':prime,
   'hotstar_ott':hotstar}
```

```python
moviedetails=pd.DataFrame(d)
moviedetails
```

```python
moviedetails.head(2)
df=moviedetails
```
