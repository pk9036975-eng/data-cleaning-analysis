```python

def comp_type(row):
    f=row['fam_count']
    p=row['no_of_psgs']

    if p==1:
        return 'solo'
    elif f==0 and p>1:
        return 'friends'
    else:
        return 'family'
    
```

```python
df['companin_type']=df.apply(comp_type,axis=1)
```

```python
df
```

```python
df['age_category']=pd.cut(df['age'],bins=[0,2,16,25,59,100],labels=['Infant','Child','Teen','Adult','Senior'],)
```

```python
royal = ['Countess', 'Lady', 'Sir', 'Don', 'Dona', 'Jonkheer']
common = ['Mr', 'Mrs', 'Miss', 'Ms', 'Master', 'Mme', 'Mlle']
military = ['Col', 'Major', 'Capt']
other_professional = ['Dr', 'Rev']
```

```python
def social_status(row):
    c=row['courtesy_title']
    if c in royal:
        return 'Royal'
    elif c in common:
        return 'Common'
    elif c in military:
        return 'Military'
    else:
        return 'Other_Professional'

df['social status']=df.apply(social_status,axis=1)
```

# conclusion

**The analysis of the Titanic dataset shows that male passengers were higher in number compared to female passengers. Pclass 3 had the largest share of passengers when compared to Pclass 1 and Pclass 2.**

**Adults formed the majority of travelers, with most passengers falling between 23 and 32 years of age. Among the survivors, adults represented the largest group.**

**Southampton recorded the highest number of ship departures compared to Queenstown and Cherbourg.**

**Most ticket fares ranged between 5 and 40, and there was no strong relationship between age and fare, except for a few passengers who paid significantly higher fares.**
